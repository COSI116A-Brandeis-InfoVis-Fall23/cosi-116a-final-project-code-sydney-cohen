```python
import numpy as np
import pandas as pd

monthly_ridership = pd.read_csv('Original Data Files/MBTA_Monthly_Ridership_by_Mode.csv')
monthly_ridership.head()
```

```python
monthly_ridership = monthly_ridership.drop(monthly_ridership.loc[monthly_ridership['mode'] != 'Rail'].index)
monthly_ridership.head()
```

```python
print(monthly_ridership.size)
monthly_ridership = monthly_ridership.dropna()
print(monthly_ridership.size)
```

```python
for date in monthly_ridership['service_date']:
     if date[0:4] != '2022':
          monthly_ridership = monthly_ridership.drop(monthly_ridership.loc[monthly_ridership['service_date'] == date].index)
monthly_ridership.head()
```

```python
monthly_ridership.to_json('monthly_rail_ridership_2022.json')
monthly_ridership.to_csv('monthly_rail_ridership_2022.csv')
```
